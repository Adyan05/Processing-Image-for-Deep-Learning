**Resizing Image**
*   Using Pillow Library



```python
from PIL import Image
img=Image.open('/content/Cat.jpg')
img_resized = img.resize((200,200))
img_resized.show()
img_resized.save('resized_cat.jpg')
```

```python
img_res = mpimg.imread('/content/resized_cat.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

**Convert RGB to Grayscale**

*   Use OpenCV Library

```python
import cv2
```

```python
img=cv2.imread('/content/Cat.jpg')
type(img)
```

```python
img.shape
grayscale_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
type(grayscale_img)
print(grayscale_img.shape)
```
